```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd

# Cargar el archivo
archivo = list(uploaded.keys())[0]
excel = pd.ExcelFile(archivo)

# Mostrar nombres de hojas
excel.sheet_names
```

```python
df = excel.parse('Stocks')
df.head(10)
```

```python
# Saltamos la primera fila vacía
df = df.iloc[1:]

# Renombramos las dos primeras columnas
df.rename(columns={df.columns[0]: 'logistica', df.columns[1]: 'unidad'}, inplace=True)

# Verificamos las nuevas columnas
df.head()
```

```python
# Derretir todas las columnas de fechas en filas
df_largo = df.melt(id_vars=['logistica', 'unidad'], var_name='fecha', value_name='valor')

# Convertir 'fecha' a tipo datetime
df_largo['fecha'] = pd.to_datetime(df_largo['fecha'])

# Ver resultado
df_largo.head(10)
```

```python
# Eliminar filas sin valor numérico
df_largo = df_largo.dropna(subset=['valor'])

# Resetear índice si querés ordenarlo
df_largo = df_largo.reset_index(drop=True)
```

```python
# Diccionario para filtrar y renombrar valores de la columna 'logistica'
stock_deseados = {
    'STOCK MAIZ': 'maiz',
    'STOCK ETANOL ANHIDRO': 'etanol',
    'STOCK DDGS SEGÚN LOGISTICA': 'ddgs'
}

# Filtrar filas
df_filtrado = df_largo[df_largo['logistica'].isin(stock_deseados.keys())].copy()

# Renombrar valores de 'logistica' y la columna en sí
df_filtrado['stock'] = df_filtrado['logistica'].map(stock_deseados)
df_filtrado = df_filtrado.drop(columns='logistica')

# Reordenar columnas
df_filtrado = df_filtrado[['fecha', 'stock', 'unidad', 'valor']]

# Ver resultado
df_filtrado.head()
```

```python
# Guardar CSV con coma como separador decimal y punto y coma como delimitador de columnas
df_filtrado.to_csv('stocks_cleaned.csv', index=False, sep=';', decimal=',')

# Descargar el archivo
from google.colab import files
files.download('stocks_cleaned.csv')
```
